```python
# Extract Data from many folders of excel files and search between them for extracting special cells

import os
import glob
from openpyxl import load_workbook

def get_digits(str1):
    c = ""
    for i in str1:
        if i.isdigit():
            c += i
    return c

path = "C:/Users/IT1/Desktop/"
extension = 'xlsx'

wb = load_workbook(path+"file.xlsx", data_only=True)
ws=wb['Sheet1']
for j in range(1,2249):
    Vmci='No data'
    Vmtn='No data'
    strcode= str(ws['A'+str(j)].value)
    for folder in os.listdir(path):
        if(get_digits(folder) == strcode or get_digits(folder) == '0'+strcode):
            os.chdir(path+folder)
            result = glob.glob('*.{}'.format(extension))
            if('i' in result[0] or 'I' in result[0]):#if mci is first
                print(j)
                wb1 = load_workbook(path+folder+'/'+result[0], read_only=True, data_only=True)
                ws1=wb1['Level']
                if(ws1['AV89'].value == 100):
                    Vmci = 0
                elif(ws1['AS95'].value != '#DIV/0!'):
                    Vmci= ws1['AS95'].value
                wb2 = load_workbook(path+folder+'/'+result[1], read_only=True, data_only=True)
                ws2=wb2['Level']
                if(ws2['AV89'].value == 100):
                    Vmtn = 0
                elif(ws2['AS95'].value != '#DIV/0!'):
                    Vmtn= ws2['AS95'].value
            else:
                wb1 = load_workbook(path+folder+'/'+result[1], read_only=True, data_only=True)
                ws1=wb1['Level']
                if(ws1['AV89'].value == 100):
                    Vmci = 0
                elif(ws1['AS95'].value != '#DIV/0!'):
                    Vmci= ws1['AS95'].value
                wb2 = load_workbook(path+folder+'/'+result[0], read_only=True, data_only=True)
                ws2=wb2['Level']
                if(ws2['AV89'].value == 100):
                    Vmtn = 0
                elif(ws2['AS95'].value != '#DIV/0!'):
                    Vmtn= ws2['AS95'].value
    ws['B'+str(j)] = Vmci
    ws['C'+str(j)] = Vmtn
                
wb.save(path+"new_file.xlsx")
```
